```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, 
        color='#FF8C00', # RGB Hex Color
        lw=3,alpha=0.5, 
        ls='-', 
        marker='o',
        ms=15,
        mfc='yellow',
        mew='2',
        mec='black') 

plt.show()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='purple', lw=2, ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,1])

plt.show()
```
