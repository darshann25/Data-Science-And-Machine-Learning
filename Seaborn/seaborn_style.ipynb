```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.set_style(style='ticks')
sns.set_context('notebook')
sns.countplot(x='sex',data=tips)
sns.despine()
```

```python
sns.lmplot('total_bill','tip',data=tips,hue='sex',palette='afmhot')
```
